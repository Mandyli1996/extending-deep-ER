```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import rcParams

sns.set(context='notebook', style='white')
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = 'times new roman'

import matplotlib 
matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20)
matplotlib.rc('axes', labelsize=20)

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
results = pd.read_excel('../notebooks/results-tidy.xlsx')

f1 = results[results.measurement == 'f1']
f1 = f1.pivot_table(index='dataset', columns='model', values='score')

datasets = ['amazon-google', 'amazon-walmart', 'dblp-scholar']
```

# MALSTM does not work well without more data

```python
columns = ['avg', 'lstm', 'malstm', 'malstm-t']
ax = f1[columns].plot.bar(figsize=(15,7),
                          color=['lightgray', 'gray', 'pink', 'red'],
                          alpha=0.75)
ax.set_xticklabels(datasets, rotation=0)
ax.set(ylim = (.5, 1), ylabel = 'f1-score')
ax.yaxis.grid(True)
ax.legend(fontsize=14)
```

# Unable to verify deepER results

```python
columns = ['published-best', 'avg']
ax = f1[columns].plot.bar(figsize=(15,7),
                          color=['gray', 'red'],
                          alpha=0.75)
ax.set_xticklabels(datasets, rotation=0)
ax.set(ylim = (.6, 1), ylabel = 'f1-score')
ax.legend(fontsize=20, bbox_to_anchor=(1,1))
ax.yaxis.grid(True)
```

# IDF tends to perform worse. LSTM > AVG. Allcomp always best.

```python
columns = ['avg', 'idf', 'lstm', 'allcomp']
ax = f1[columns].plot.bar(figsize=(15,7), alpha=0.75)
ax.set_xticklabels(datasets, rotation=0)
ax.set(ylim = (.6, 1), ylabel = 'f1-score')
ax.legend(fontsize=20, bbox_to_anchor=(1,1))
ax.yaxis.grid(True)
```

# allcomp-allsim > allsim. more similarity metrics don't make much difference.

```python
columns = ['avg', 'avg-allsim', 'lstm', 'lstm-allsim', 'allcomp-allsim', 'allcomp-allsim-t']
ax = f1[columns].plot.bar(figsize=(15,7), alpha=0.75)
ax.set_xticklabels(datasets, rotation=0)
ax.set(ylim = (.6, 1), ylabel = 'f1-score')
ax.legend(fontsize=20, bbox_to_anchor=(1,1))
ax.yaxis.grid(True)
```

# avg-t > avg. lstm-t mostly better than lstm. avg-t best option in two cases; third case not behind by much.

```python
columns = ['avg', 'avg-t', 'lstm', 'lstm-t']
ax = f1[columns].plot.bar(figsize=(15,7), alpha=0.75)
ax.set_xticklabels(datasets, rotation=0)
ax.set(ylim = (.6, 1), ylabel = 'f1-score')
ax.legend(fontsize=20, bbox_to_anchor=(1,1))
ax.yaxis.grid(True)
```

# Full vocabulary almost always worse. Pre-processsing matters.

```python
columns = ['avg', 'avg-full', 'lstm', 'lstm-full']
ax = f1[columns].plot.bar(figsize=(15,7), alpha=0.75)
ax.set_xticklabels(datasets, rotation=0)
ax.set(ylim = (.6, 1), ylabel = 'f1-score')
ax.legend(fontsize=20, bbox_to_anchor=(1,1))
ax.yaxis.grid(True)
```

# Adding num or num-null always better, but not clear which one.

```python
columns = ['avg', 'avg-num', 'avg-num-null', 'lstm', 'lstm-num', 'lstm-num-null']
ax = f1[columns].plot.bar(figsize=(15,7), alpha=0.75)
ax.set_xticklabels(datasets, rotation=0)
ax.set(ylim = (.6, 1))
ax.yaxis.grid(True)
```

# When using averaging, trainable embedding always better. Best is always one of num-t or num-null-t. Additional sim metrics not much help.

```python
columns = ['avg', 'avg-t', 'avg-num-t', 'avg-num-null-t', 'avg-num-null-allsim-t']
ax = f1[columns].plot.bar(figsize=(15,7), alpha=0.75)
ax.set_xticklabels(datasets, rotation=0)
ax.set(ylim = (.6, 1), ylabel = 'f1-score')
ax.legend(fontsize=20, bbox_to_anchor=(1,1))
ax.yaxis.grid(True)
```

# When using LSTM composition, num-null always helps.

```python
columns = ['lstm', 'lstm-allsim', 'lstm-t', 'lstm-full', 'lstm-num', 'lstm-num-null']
ax = f1[columns].plot.bar(figsize=(15,7), alpha=0.75)
ax.set_xticklabels(datasets, rotation=0)
ax.set(ylim = (.6, 1), ylabel = 'f1-score')
ax.legend(fontsize=20, bbox_to_anchor=(1,1))
ax.yaxis.grid(True)
```

# When using AVG composition, num-null always helps. Trainable embeddings always helps. Full vocab always hurts.

```python
columns = ['avg', 'avg-allsim', 'avg-t', 'avg-full', 'avg-num', 'avg-num-null']
ax = f1[columns].plot.bar(figsize=(15,7), alpha=0.75)
ax.set_xticklabels(datasets, rotation=0)
ax.set(ylim = (.6, 1), ylabel = 'f1-score')
ax.legend(fontsize=20, bbox_to_anchor=(1,1))
ax.yaxis.grid(True)
```
